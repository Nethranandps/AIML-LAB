Write a program to implement k-Nearest Neighbour algorithm to classify the iris data 
set. Print both correct and wrong predictions.


```python
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from collections import Counter
import numpy as np
```

```python
iris=datasets.load_iris()
```

```python
iris
```

```python
x=iris.data
y=iris.target
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.3,random_state=48)
```

```python
def euclidean_distance(x1, x2):
    return np.sqrt(np.sum(x1-x2)**2)

class KNN:
    def __init__(self,k=3):
        self.k=k
    
    def fit(self,x,y):
        self.x_train=x
        self.y_train=y
        
    def predict(self,x):
        y_pred=[self._predict(x) for x in x]
        return np.array(y_pred)
    
    def _predict(self,x):
        #compute distances b/w x and all samples in the training set
        distances=[euclidean_distance(x,x_train) for x_train in self.x_train]
        #sort by distance and return indices of the first k samples
        k_indices=np.argsort(distances)[:self.k]
        #extract the labels of the k nearest neighbour training samples
        k_nearest_labels=[self.y_train[i] for i in k_indices]
        #return the most common class labels
        most_common=Counter(k_nearest_labels).most_common(1)
        return most_common[0][0]

#initialise and fit the model
k=5
clf=KNN(k=k)
clf.fit(x_train, y_train)
predictions=clf.predict(x_test)

correct_predictions=sum(predictions==y_test)
print(f"Correct predictions:{correct_predictions}/{len(y_test)}")

wrong_indices=np.where(predictions!=y_test)[0]
print("Wrong prediction ")
for index in wrong_indices:
        print(f"Predicted: {iris.target_names[predictions[index]]}, Actual: {iris.target_names[y_test[index]]}")
              
```
